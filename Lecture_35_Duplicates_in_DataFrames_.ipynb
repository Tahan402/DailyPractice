```python
import numpy as np
import pandas as pd
```

```python
dframe = pd.DataFrame({'key1':['A']*2 + ['B']*3 , 
                       'key2':[2,2,2,3,3]})

dframe
```

```python
dframe.duplicated()
```

```python
dframe.drop_duplicates() # for drop duplicates row
```

```python
dframe.drop_duplicates(['key1']) #for drop duplicates column
```
