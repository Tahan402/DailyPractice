```python
import numpy as np
import pandas as pd
```

```python
dframe1 = pd.DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})
```

```python
dframe1
```

```python
dframe2 = pd.DataFrame({'key':['Q','Y','Z'],'data_set_2':[1,2,3]})
```

```python
dframe2
```

```python
pd.merge(dframe1,dframe2) # merge the two dataframe
```

```python
pd.merge(dframe1,dframe2,on='key') # merge by specific columns
```

```python
pd.merge(dframe1,dframe2,on='key',how='left') # which dataframe to used all the values of the column
```

```python
dframe2
```

```python
dframe1
```

```python
pd.merge(dframe1,dframe2,on='key',how='right')
```

```python
pd.merge(dframe1,dframe2,on='key', how='outer') #right than sql dataset merge thing
```

```python
df_left = pd.DataFrame({'key1':['SF','SF','LA'],'key2':['one','two','one'],'left_data':[10,20,30]})
```

```python
df_right=pd.DataFrame({'key1':['SF','SF','LA','LA'],
                       'key2':['one','one','one','two'],
                       'right_data':[40,50,60,70]})
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_left,df_right,on=['key1','key2'],how='outer')
```

```python
pd.merge(df_left,df_right,on='key1') # pandas automaticaly seprate column for you
```

```python
pd.merge(df_left,df_right,on='key1',suffixes=('_lefty','_righty'))
```
