```python
import numpy as np
import pandas as pd
```

```python
dframe_wine = pd.read_csv('/content/drive/My Drive/Data_Set_python/winequality-red.csv', sep = ';')
```

```python
dframe_wine.head()
```

```python
def ranker(df):
  df['alc_content_rank'] = np.arange(len(df)) +1
  return df
```

```python
dframe_wine.sort_values('alcohol' ,ascending = False, inplace= True)
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
dframe_wine.head()
```

```python
num_of_qual =dframe_wine['quality'].value_counts()
```
